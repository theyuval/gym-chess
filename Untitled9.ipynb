```python
# Assume df is your DataFrame and it has a 'sub_area' column.
# For example:
# df = pd.read_csv("train.csv")

# --- 1. Group Rare Sub Areas ---
# Set a threshold: for instance, sub areas that appear less than 50 times will be grouped as "Other".
rare_threshold = 10
sub_area_counts = df['sub_area'].value_counts()
# Identify rare
rare_sub_areas = sub_area_counts[sub_area_counts < rare_threshold].index
print("rare sub areas:",rare_sub_areas )
# Replace rare sub areas with the string 'Other'
df['sub_area'] = df['sub_area'].replace(rare_sub_areas, 'Other')

# Optionally, verify the grouping:
print("Sub-area counts after grouping rare values:")
print(df['sub_area'].value_counts())

# --- 2. Impute Missing Values Based on Sub-Area Median ---
# Select numeric columns (adjust as needed if you want to impute only a subset)
num_cols = df.select_dtypes(include=['float64', 'int64']).columns
cat_cols = df.select_dtypes(include=['object']).columns
# For each numeric column with missing values, fill NaNs with the median value for that sub_area.
for col in num_cols:
    if df[col].isnull().sum() > 0:
        df[col] = df.groupby('sub_area')[col].transform(lambda x: x.fillna(x.median()))

for col in cat_cols:
    if df[col].isnull().sum() > 0:
        df[col] = df.groupby('sub_area')[col].transform(lambda x: x.fillna(x.mode()[0]))

# Optionally, check that missing values are now imputed:
print("Missing values per column after imputation:")
print(df.isnull().sum())
```

```python
def calculate_space_averages(df, ranges=[500, 1000, 1500, 2000, 3000, 5000]):
    """
    Calculate average space for office and TRC for each range
    """
    for range_val in ranges:
        # Calculate office average space
        df[f'office_avg_space_{range_val+1}'] = (
            df[f'office_sqm_{range_val}'].div(df[f'office_count_{range_val}'])
                .fillna(0)
                .replace([np.inf, -np.inf], 0)
        )

        # Calculate TRC average space
        df[f'trc_avg_space_{range_val+1}'] = (
            df[f'trc_sqm_{range_val}'].div(df[f'trc_count_{range_val}'])
                .fillna(0)
                .replace([np.inf, -np.inf], 0)
        )

    return df


# Apply the function to your DataFrame
df = calculate_space_averages(df)
```

```python

def drop_high_corr_features(X, corr_threshold=0.90):
    # Compute the correlation matrix and convert to a condensed distance matrix
    corr_matrix = X.corr().abs()
    dist_matrix = 1 - corr_matrix  # Convert to "distance"
    dist_matrix = dist_matrix.clip(lower=0) #avoid negative values

    condensed_dist = squareform(dist_matrix)

    # Perform hierarchical clustering
    linkage_matrix = linkage(condensed_dist, method='average')
    cluster_labels = fcluster(linkage_matrix, t=1 - corr_threshold, criterion='distance')

    # Group features by clusters
    cluster_map = {}
    for i, label in enumerate(cluster_labels):
        cluster_map.setdefault(label, []).append(corr_matrix.columns[i])

    # Retain one representative feature per cluster
    retained_features = []
    for cluster, features in cluster_map.items():
        if len(features) > 1:
            # Retain the feature most correlated with the target variable
            representative = features[0]  # Default to the first feature
            retained_features.append(representative)
        else:
            retained_features.extend(features)

    to_drop = [col for col in X.columns if col not in retained_features
              ]# and 'pca' not in col]  apparently not helping to improve prediction
    return to_drop, retained_features
all_columns = X_investment.columns
excluded_cols = ["_particular_1m_2m_3m_missing","_particular_1m_2m_3m_magic"]  # add any columns you want to exclude
num_cols = [
    col
    for col in all_columns
    if df[col].dtype in ["int64", "float64"] and col not in excluded_cols
]

to_drop, retained_features = drop_high_corr_features(X_investment[num_cols], corr_threshold=0.95)
X_investment.drop(columns=to_drop, inplace=True)
X_personal.drop(columns=to_drop, inplace=True)
print("columns dropped:", to_drop)
```

```python
import matplotlib.pyplot as plt
import scipy.stats as stats

# --- 1. Compute Residuals ---
# Assuming df_test_combined already contains the columns "actual" and "prediction"
df_test_combined['residual'] = df_test_combined['actual'] - df_test_combined['prediction']

# --- 2. Plot Residual vs. Fitted (Predicted) Values ---
plt.figure(figsize=(10, 6))
# If you have a location column (e.g., 'sub_area'), use it as the hue for additional insight.
if 'sub_area' in df_test_combined.columns:
    sns.scatterplot(x="prediction", y="residual", hue="sub_area", data=df_test_combined, palette="viridis")
else:
    sns.scatterplot(x="prediction", y="residual", data=df_test_combined, color="blue")
plt.axhline(0, color="red", linestyle="--")
plt.xlabel("Fitted Values (Predicted Price)")
plt.ylabel("Residual (Actual - Predicted)")
plt.title("Residual vs. Fitted Values")
plt.show()

# --- 3. Plot Histogram and Q-Q Plot of Residuals ---
plt.figure(figsize=(8, 6))
sns.histplot(df_test_combined['residual'], kde=True, color="skyblue")
plt.title("Distribution of Residuals")
plt.xlabel("Residual")
plt.ylabel("Frequency")
plt.show()

plt.figure(figsize=(8, 6))
stats.probplot(df_test_combined['residual'], dist="norm", plot=plt)
plt.title("Q-Q Plot of Residuals")
plt.show()

# --- 4. Identify and Explore Outliers ---
# Here we define outliers as observations where the absolute residual is greater than 2 standard deviations.
resid_std = df_test_combined['residual'].std()
outlier_threshold = 2 * resid_std

# Filter for outliers:
outliers = df_test_combined[np.abs(df_test_combined['residual']) > outlier_threshold]

print("Outliers where |residual| > {:.2f}:".format(outlier_threshold))
# If a location column exists, include it in the printed table.
if 'sub_area' in outliers.columns:
    print(outliers[['actual', 'prediction', 'residual', 'sub_area']])
else:
    print(outliers[['actual', 'prediction', 'residual']])

# (Optional) If you want to summarize errors by location (for example, by "sub_area"):
if 'sub_area' in df_test_combined.columns:
    error_summary = df_test_combined.groupby('sub_area')['residual'].agg(['mean', 'std', 'count'])
    print("\nError summary by sub_area:")
    print(error_summary)
```
